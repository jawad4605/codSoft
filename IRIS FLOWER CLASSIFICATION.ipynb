```python
import numpy as np
# Define a function to plot confusion matrix
def plot_confusion_matrix(y_true, y_pred, model_name):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 4))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=np.unique(y_true),
                yticklabels=np.unique(y_true))
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title(f'Confusion Matrix for {model_name}')
    plt.show()

# Plot confusion matrix for each model
models = {
    'Logistic Regression': y_pred_logreg,
    'Decision Tree': y_pred_decision_tree,
    'K-Nearest Neighbors': y_pred_knn,
    'Support Vector Machine': y_pred_svm
}

for model_name, y_pred in models.items():
    plot_confusion_matrix(y_test, y_pred, model_name)
```

```python
# Initialize lists to store the metrics
model_names = []
accuracies = []
precisions = []
recalls = []
f1_scores = []

# Calculate and store the metrics for each model
for model_name, y_pred in models.items():
    accuracy = accuracy_score(y_test, y_pred)
    cm = confusion_matrix(y_test, y_pred)
    precision = cm[1, 1] / (cm[1, 1] + cm[0, 1])  # Precision for class 1
    recall = cm[1, 1] / (cm[1, 1] + cm[1, 0])     # Recall for class 1
    f1 = 2 * (precision * recall) / (precision + recall)  # F1-score for class 1
    
    model_names.append(model_name)
    accuracies.append(accuracy)
    precisions.append(precision)
    recalls.append(recall)
    f1_scores.append(f1)

# Create a DataFrame to display the consolidated report
report_df = pd.DataFrame({
    'Model': model_names,
    'Accuracy': accuracies,
    'Precision': precisions,
    'Recall': recalls,
    'F1 Score': f1_scores
})

# Display the consolidated report
print("Consolidated Report for All Models:")
print(report_df)
```
